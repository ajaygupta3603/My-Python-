1. Which team won the most matches?



```python
match_wins = df['match_winner'].value_counts()
```

```python
sns.barplot(x = match_wins.values, y = match_wins.index, palette = 'rainbow')
plt.title("Most match win by team")
plt.xlabel("Numer of matches win")
plt.ylabel("Teams")
```

```python
#2 - Top Scorers

topscorer = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending= False).head()
```

```python
sns.barplot(y = topscorer.index, x = topscorer.values)
plt.grid(axis= 'x', linestyle = '--')
```

```python
#3 10 Best Bowling Figures

df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x: x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
```

```python
bestbowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending= False).head(10)
```

```python
bestbowlers.plot(kind = 'barh')
plt.grid(axis = 'x',color='g', linestyle='--')
```
