```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np 
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(x= df['sex'], hue= df['smoker'] )
```

```python
sns.barplot(df,x ='sex', y = 'total_bill', hue = 'day', estimator= np.sum, palette = 'rainbow')
```

```python
sns.boxplot(df, x= 'sex', y= 'total_bill', hue = 'smoker', palette='rainbow')
```

```python
sns.violinplot(df,x = 'sex', y= 'total_bill', hue= 'day')
```

```python
sns.stripplot(df,x = 'sex', y= 'tip')
```

```python
sns.swarmplot(df,x = 'sex', y= 'tip')
```
