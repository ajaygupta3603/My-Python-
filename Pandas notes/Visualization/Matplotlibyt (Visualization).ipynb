```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x**3
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('(X-axis)')
plt.ylabel('(Y-axis)')
plt.title('(Basic Matplotlib diagram)')
```

```python
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x*2,y)
plt.subplot(2,2,3)
plt.plot(x**3,y)
```

```python
fig = plt.figure(figsize= (5,5),dpi = 100)
a1 = fig.add_axes([.1,.1,.6,.6])
a1.plot(x,y)
a2 = fig.add_axes([.2,.2,.2,.2])
a2.plot(x,y)

```

```python

fig = plt.figure()
axis0 = fig.add_axes([.1,.1,.2,.2])
plt.savefig("Basic Plot.jpg")
```
